```python
import csv
f = open("seoul.csv", encoding="utf-8")
data = csv.reader(f)
next(data)
max_temp = 0
max_temp_date = ''
for row in data:
    if row[-1] != '':
        if int(row[0].split("-")[0]) < 1995 and float(row[-1]) > max_temp:
            max_temp = float(row[-1])
            max_temp_date = row[0]

print(max_temp, max_temp_date)     
```

```python
import csv
f = open("seoul.csv", encoding="utf-8")
data = csv.reader(f)
next(data)
temperature = []
for row in data:
    if (row[-1] != '' 
        and (row[0].split('-')[1] == '11' and row[0].split('-')[2] == '24')):
        temperature.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.plot(temperature)
plt.show()
```

```python
# 역대 서울 기온 중 37 이상이었던 날의 날짜와 최고 기온을 출력하시오.
import csv
f = open("seoul.csv", encoding="utf-8")
data = csv.reader(f)
next(data)
max_temp = 0
max_temp_date = ''
for row in data:
    if row[-1] != '':
        if float(row[-1]) > 37 and float(row[-1]) > max_temp:
            max_temp = float(row[-1])
            max_temp_date = row[0]

print(max_temp, max_temp_date)  
```
