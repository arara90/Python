```python
#2.함수를 사용해서 구현하기
def main():
    print( "\n[[[[[[[ 시 작 ]]]]]]]")
    n1 = input("첫번째 수 : ")
    n2 = input("두번째 수 : ")
    
    print( "원하는 연산 번호를 입력하세요")
    print( "1. +" )
    print( "2. -" )
    print( "3. *" )
    print( "4. /" )
    cal = input(">> 입력 : ")

def calc(num1, num2, cal1):
    if cal1 == 1 :
        return num1 + num2 
    elif cal1 == 2 :
        return num1 - num2 
    elif cal1 == 3 :
        return num1 * num2 
    elif cal1 == 4 :
        return num1 / num2 
    else :
        x = input("잘못 입력했습니다. 종료하려면 y, 다시 하려면 any : ")
        if(x=='y'):
            return 0
       
while(1):
    main()  #메인
    try:
        num1 = int(n1)
        num2 = int(n2)
        cal1 = int(cal)
    except:
        x = input("★잘못 입력했습니다. 종료하려면 y, 다시 하려면 any : ")
        if(x=='y'):
            break;
    else:
        res = calc(num1, num2, cal1 )
        if ( res ) :
            print(">> 결과 :",res)
        else:
            break;
```

```python
#3.클래스를 이용해서 구현하기
## 3_1. 클래스 선언 부분 ##
class  MyCalculation :
    num1=0
    num2=0
    sel=0
    def add(self) :
        return (self.num1) + (self.num2) 
    def sub(self) :
        return (self.num1) - (self.num2)
    def mul(self) :
        return (self.num1) * (self.num2) 
    def div(self) :
        return (self.num1) / (self.num2)
    
    def cal(self):
        if self.sel == 1 :
            return self.add() 
        elif self.sel == 2 :
            return self.sub()
        elif self.sel == 3 :
            return self.mul() 
        elif self.sel == 4 :
            return self.div()
        else :
            x = input("[in Class]잘못 입력했습니다. 종료하려면 y, 다시 하려면 any : ")
            if(x=='y'):
                return 0

    
```
