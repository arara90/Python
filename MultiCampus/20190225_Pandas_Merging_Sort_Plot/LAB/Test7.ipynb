```python
# numpy, pandas, matplotlib
# 그래프, 도식화

import matplotlib.pyplot as plt
plt.figure()
plt.plot([1,2,3,4,4,3,2,5,6,7,7,8,9,10])
plt.show()
```

```python
import numpy as np
x = np.arange(0,12,0.01)
y = np.sin(x)
y1 = np.cos(x)
plt.figure(figsize=(10,2))
plt.plot(x,y)
plt.plot(x, y1)
plt.grid()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("x-y title")
plt.show()
```

```python
import numpy as np
x = np.arange(0,12,0.01)
y = np.sin(x)
y1 = np.cos(x)
plt.figure(figsize=(10,4))
plt.plot(x,y, label="sin") # label 출력 그래프 구별 목록 표시
plt.plot(x, y1, label="cos") # label 출력 그래프 구별 목록 표시
plt.legend() # 출력 그래프 구별 목록
plt.grid()
plt.xlabel("x axis") # x 타이틀
plt.ylabel("y axis") # y 타이틀
plt.title("x-y title") # main 타이들
plt.ylim(-1.2, 1.2) # y축확대
plt.xlim(0, np.pi) # x축확대
plt.show()
```

```python
t = np.arange(0, 5, 0.5)
plt.figure(figsize=(10,6))
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'b>')
plt.plot(t, t**3, 'gx')
plt.plot(t, t**2.5, color='green', linestyle='dashed')
plt.plot(t, t**2.5, color='green', linestyle='dashed', marker='o')
plt.plot(t, t**2.5, color='green', linestyle='dashed', marker='o', 
         markerfacecolor='red')
plt.plot(t, t**2.5, color='green', linestyle='dashed', marker='o', 
         markerfacecolor='red', markersize=20)
plt.show()
```
