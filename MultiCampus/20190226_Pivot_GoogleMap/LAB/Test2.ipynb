```python
import pandas as pd
import numpy as np

crime_seoul = pd.read_csv("crime_Seoul2.csv", encoding='utf-8',  
                         index_col=0)
#crime_seoul['구별']
crime_seoul
crime_pivot = pd.pivot_table(crime_seoul, index='구별', aggfunc=np.sum)
crime_pivot.head()

# 컬럼 추가(5)
# 강간검거율, 강도검거율, 살인검거율, 절도검거율, 폭력검거율
# 강도거거율 = 강도 검거 / 강도 발생 * 100
crime_pivot['강간검거율'] = crime_pivot['강간 검거'] / crime_pivot['강간 발생'] * 100
crime_pivot['강도검거율'] = crime_pivot['강도 검거'] / crime_pivot['강도 발생'] * 100
crime_pivot['살인검거율'] = crime_pivot['살인 검거'] / crime_pivot['살인 발생'] * 100
crime_pivot['절도검거율'] = crime_pivot['절도 검거'] / crime_pivot['절도 발생'] * 100
crime_pivot['폭력검거율'] = crime_pivot['폭력 검거'] / crime_pivot['폭력 발생'] * 100
```

```python
del crime_pivot['강간 검거']
del crime_pivot['강도 검거']
del crime_pivot['살인 검거']
del crime_pivot['절도 검거']
del crime_pivot['폭력 검거']
crime_pivot.head()
```

```python
cols = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
for s in cols:
    crime_pivot.loc[crime_pivot[s] > 100, s] = 100
crime_pivot.head()
```
