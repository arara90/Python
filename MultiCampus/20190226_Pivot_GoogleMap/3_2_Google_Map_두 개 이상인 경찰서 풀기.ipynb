```python
import pandas as pd
import numpy as np
#!pip install googlemaps
import googlemaps

gmaps_key = "AIzaSyA0WavkAEvRu8N9ZqDUKyo4tAisTY8Egj4"
gmaps = googlemaps.Client(key=gmaps_key)

crime = pd.read_csv("Data/crime_Seoul2.csv",encoding = 'utf-8', index_col = 0)

####1. 경찰서명 받아오기####
sttns = []
for name in crime['관서명']:
    sttns.append('서울' + name[:-1] + '경찰서')
print(sttns)
```

```python
#연습
tmp2=gmaps.geocode('서울중부경찰서',language='ko')
tmp2[0].get("formatted_address")
tmp3 = tmp2[0].get("geometry")
#위치정보 받기
#print(tmp3['location']['lat'], tmp3['location']['lng'])
lat = tmp3['location']['lat']
lng = tmp3['location']['lng']
```

```python
####2. 위치 정보 뽑아오기 ####
# (1) 각각 경찰서의 정보를 gamps에서 가져와 tmp에 저장 ( list고 내부는 섞여있음)
tmp = []
fin_res = []
for sttn in sttns:
    tmp.append(gmaps.geocode(sttn,language='ko'))

# (2) 다 풀어서 list에 dict로 저장 
for i in range(0,len(tmp)):
    if len(tmp[i])>1:
        for item in tmp[i]:
            #print(type(item))
            fin_res.append(item)
    else:
        #print(type(tmp[i][0]))
        fin_res.append(tmp[i][0])

# (3) 위도, 경도 저장
lats = []
lngs = []
for res in fin_res:
    lats.append(res.get('geometry')['location']['lat'])
    lngs.append(res.get('geometry')['location']['lng'])
    
print(len(fin_res))
```

```python
#3.지도상에 그리기
# (1) 하기전에 설치 !pip install folium
import folium as f

# (2) 지도 focus 위치 정해서 저장
#fm = f.Map(location=[lats[0],lngs[0]], zoom_start=100)
fm = f.Map(location=[37.567,127],zoom_start=12) #적도, 위도 한가운데에 zoom-in 된다.

# (3) 아이콘 찍기
for i in range(0, len(fin_res)):
    f.Marker([lats[i],lngs[i]] ,icon = f.Icon(icon='cloud')).add_to(fm)
    
# (4) 출력
fm
#두 개의 용산 경찰서 찍힘! ( 용산경찰서원효지구대, 용산경찰서 )
```
