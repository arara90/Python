```python
####2. 위치 정보 뽑아오기 ####
# (1) 각각 경찰서의 정보를 gamps에서 가져와 tmp에 저장 ( list고 내부는 섞여있음)
tmp = []
for sttn in sttns:
    tmp.append(gmaps.geocode(sttn,language='ko'))

#print(type(tmp[0])) #list
##왜 list로 저장됐을까? -> ex) 용산경찰서는 2개가 찍혀! 하나의 검색어에도 두개의 결과가 나오는 경우가 발생해!
## 이것도 풀어줘야겠다~  아래 코드에서 두개짜리로 나오는거 확인하자 => 3_2에 구현
#for i in range(0,len(tmp)):
#     print(i, " : ", type(tmp[i]), "len : " , len(tmp[i]))
#     if len(tmp[i])>1:
#         print(tmp[i])
# 

lats = []
lngs = []
for i in range(0, len(tmp)):
   # print(type(tmp[i]))
    li = tmp[i]
    lats.append(li[0].get('geometry')['location']['lat'])
    lngs.append(li[0].get('geometry')['location']['lng'])
  
print(len(lats), len(lngs))
#for item in tmp:
#    print(item
```
