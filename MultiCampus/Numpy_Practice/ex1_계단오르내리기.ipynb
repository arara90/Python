```python
#발자취
import numpy as np
import random
from matplotlib import pyplot as pl 
position = 0
walk =  [position]
steps = 1000 #1000번 실행

for i in range(steps):
    step = 1 if random.randint(0,1) else -1 
    position += step
    walk.append(position)
    
nsteps = 100
draws = np.random.randint(0, 2, size = nsteps) #size는 난수의 숫자 즉, 0-2사이에서 nstep개만큼 구하라
steps = np.where(draws > 0 , 1, -1)
#print(steps)
walk = steps.cumsum()
#print(walk)

#1. 제일 높은 곳
print(walk.max())
#2. 제일 낮은 곳
print(walk.min())
# 최초로 10칸 떨어지기까지의 시행횟구하기
x = ''
y = ''
tf = np.abs(walk) == 5

for i in range(0,len(tf)):
    if tf[i] :
        x =  i
        y = walk[i]
        break
        
fig = pl.figure()
ax = fig.add_subplot(1,1,1)

pl.plot(walk)
pl.plot(np.zeros(nsteps) + 5,'r-')
pl.plot(np.zeros(nsteps) - 5,'b--')

if (x is not None) :
    pl.plot(x,y,'yo')
    ax.annotate(y #label
                , xy = (x,y) #위치
                , xytext = (x+1,y+0.5) #text 위치 (xy기준으로 + -)
                #, arrowprops = dict(facecolor='black')
                #,horizontalalignment ='left', verticalalignment='top'
               )
```
