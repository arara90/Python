```python
import pandas as pd
import numpy as np

#Series(1차원)

#DataFrame(2차원)
data = [1,2,3,4]
df = pd.DataFrame(data)
print(df)

data2 =[['item1',1],['item2',2],['item3',3]]
np.array(data2).shape
df2 = pd.DataFrame(data2,columns=['name','qty'], index=['First', 'Second', 'Third'] )
print(df2)

##########################################################################################
df3 = pd.read_csv("seoul.csv", encoding = "utf-8")
#df.head()
#df.tail
#df['지점']
#df.columns
#df.columns[0]
#df.columns[0:2]
#df[1:2]
#df.head()
#df.iloc[0]

####loc or iloc
#      loc or iloc을 쓰지않으면  행.열 구분 모호할 수도 있으니 사용하는 것을 추천하고, 
#      행열을 동시에 인덱싱할 수 있는 것이 특징이다.
#       iloc은 numpy의 array 인덱싱 방식으로 가져옴. 즉, 칼럼명이 아니라 index숫자로서 가져올때

print (df3.loc[2:4] ) 
print (df3.iloc[2:4] ) 
print(df3[2:4])        #row or column만 가능 / default는 row자르기

######loc과 iloc의 차이#
#print(df3[[0,2,4], 0:3])  #Error
#print(df3.loc[0:5, 0:3]) #Error
print(df3.iloc[0:5, 0:3])

#print(df3.iloc[0:5,['날짜','지점']]) #Error
print(df3.loc[0:5,['날짜','지점']]) 

print(df3.iloc[[0,2,4], 0:3])
print(df3[['날짜','지점']]) #column
print(df3[0:1])#row
```
