```python
import csv
f = open("village_population.csv", encoding="utf-8")
data = csv.reader(f)
next(data)
result = []
for row in data:
    if '송도1동' in row[0]:
        for i in range(3, len(row)):
            result.append(int(row[i]))            
print(result)
import matplotlib.pyplot as plt
plt.plot(result)
plt.show()

plt.hist(result)
plt.show()


```
