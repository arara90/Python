```python
import  matplotlib.pyplot as plt
import csv

f = open("village_population.csv", encoding='utf-8')
data = csv.reader(f)
next(data)
li = []

for row in data:
    if (row[0].find("송도1동") > -1):
        for i in range(3,len(row)):
            li.append(int(row[i]))
print(li)

plt.title("Songdo 1Dong")

plt.xlabel("연령")
plt.ylabel("인구")

plt.plot(li)
plt.show()

plt.hist(li, bins = 15) #분포
plt.show()
        
    
    
```
